```python
import networkx as nx
import matplotlib.pyplot as plt

# Crear el grafo
G = nx.DiGraph()

# Definir los estados como nodos
# Estado representado como (lobo, cabra, col, granjero)
# "L" = izquierda del río, "R" = derecha del río

states = {
    'Inicio': ('L', 'L', 'L', 'L'),
    'A: Cruza Cabra': ('L', 'R', 'L', 'R'),
    'B: Regresa Solo': ('L', 'L', 'L', 'L'),
    'C: Cruza Lobo': ('R', 'L', 'L', 'R'),
    'D: Regresa Cabra': ('L', 'L', 'L', 'L'),
    'E: Cruza Col': ('R', 'L', 'R', 'R'),
    'F: Regresa Solo': ('L', 'L', 'L', 'L'),
    'G: Cruza Cabra Final': ('R', 'R', 'L', 'R'),
}

# Agregar nodos
for state in states:
    G.add_node(state)

# Definir las transiciones válidas o aristas
transitions = [
    ('Inicio', 'A: Cruza Cabra'),
    ('A: Cruza Cabra', 'B: Regresa Solo'),
    ('B: Regresa Solo', 'C: Cruza Lobo'),
    ('C: Cruza Lobo', 'D: Regresa Cabra'),
    ('D: Regresa Cabra', 'E: Cruza Col'),
    ('E: Cruza Col', 'F: Regresa Solo'),
    ('F: Regresa Solo', 'G: Cruza Cabra Final')
]

# Agregar las aristas al grafo
G.add_edges_from(transitions)

# Definir las posiciones para evitar que se ordene aleatoriamente
pos = {
    'Inicio': (0, 0),
    'A: Cruza Cabra': (1, 0),
    'B: Regresa Solo': (2, 0),
    'C: Cruza Lobo': (3, 0),
    'D: Regresa Cabra': (4, 0),
    'E: Cruza Col': (5, 0),
    'F: Regresa Solo': (6, 0),
    'G: Cruza Cabra Final': (7, 0)
}

# Dibujar el grafo
plt.figure(figsize=(14,6))
nx.draw(
    G, pos,
    with_labels=True,
    node_size=3000,
    node_color='skyblue',
    edgecolors='black',  # Agrego bordes a los nodos
    font_size=9,
    arrows=True,
    arrowstyle='-|>',
    arrowsize=20
)
plt.title("Problema del Granjero, Lobo, Cabra y Col", fontsize=15)
plt.show()
```
