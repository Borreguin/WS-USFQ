```python
import networkx as nx
import matplotlib.pyplot as plt
```

```python
#Crear el grafo
G = nx.DiGraph()

#Defino cada estado como nodos
#El estado representado: lobo, cabra, col, granjero
#La derecha = "R", la izquierda = "L", esto representa el lado del río en el que están

states = {
    'Inicio': ('L', 'L', 'L', 'L'),
    'A: Cruza Cabra con granjero': ('L', 'R', 'L', 'R'),
    'B: Regresa solo granjero': ('L', 'L', 'L', 'L'),
    'C: Cruza Lobo con granjero': ('R', 'L', 'L', 'R'),
    'D: Regresa cabra con granjero': ('L', 'L', 'L', 'L'),
    'E: Cruza Col': ('R', 'L', 'R', 'R'),
    'F: Regresa solo otra vez': ('L', 'L', 'L', 'L'),
    'G: Cruza cabra final': ('R', 'R', 'L', 'R'),
}

#Agregar nodos

for state in states:
    G.add_node(state)

# Definir las transiciones válidas o aristas
transitions = [
    ('Inicio', 'A: Cruza Cabra'),
    ('A: Cruza Cabra', 'B: Regresa solo'),
    ('B: Regresa solo', 'C: Cruza Lobo'),
    ('C: Cruza Lobo', 'Regresa cabra'),
    ('D: Regresa cabra', 'E: Cruza Col'),
    ('E: Cruza Col', 'F: Regresa solo otra vez'),
    ('F: Regresa solo otra vez', 'G: Cruza cabra final'),
]

# Agregar las aristas
G.add_edges_from(transitions)

# Dibujar el grafo
plt.figure(figsize=(12,8))
pos = nx.spring_layout(G)
nx.draw(G, pos, with_labels=True, node_size=3000, node_color='skyblue', font_size=10, arrows=True)
plt.title("Problema del Granjero, Lobo, Cabra y Col", fontsize=15)
plt.show()
```
